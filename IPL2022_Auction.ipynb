```python
### null values in the datasets
ipl.isnull().sum()
```

```python
ipl[ipl['Cost IN $ (000)'].isnull()]
```

```python
ipl['Cost IN $ (000)']=ipl['Cost IN $ (000)'].fillna(0)
ipl['COST IN ₹ (CR.)']=ipl['COST IN ₹ (CR.)'].fillna(0)
```

```python
ipl[ipl['2021 Squad'].isnull()]
```

```python
ipl['2021 Squad']=ipl['2021 Squad'].fillna('Not Participated')
```

```python
ipl.isnull().sum()
```

```python
ipl
```

```python
teams=ipl[ipl['COST IN ₹ (CR.)']>0]['Team'].unique()
teams
```

```python
ipl['status']=ipl['Team'].replace(teams,'sold')
```

```python
ipl[ipl['Player'].duplicated(keep=False)]
```

```python
# no. of players participated in ipl2022
ipl['Player'].count()
```

```python
#how many types of players participated?
types=ipl.groupby('TYPE')['TYPE'].count()
# types.shape[0]
types
# type(types)
```

```python
#how many types of players participated?
## Method---2
typ=ipl['TYPE'].value_counts()
typ
typ.reset_index(name='Count')
# type(typ)
```

```python
plt.pie(typ.values,labels=typ.index,labeldistance=1.1,autopct='%1.2f%%',shadow=False,startangle=60)
plt.title("Role of players Participated",fontsize=15)
plt.show()
```

```python
#players sold and unsold using a bar graph

plt.figure(figsize=(6,3))
fig=sns.countplot(ipl['status'],palette=['Red','Blue'])
plt.ylabel("Status")
plt.xlabel('No. of players')
plt.title("Sold vs Unsold",fontsize=15)
plt.show()

# cnt=ipl.groupby('status')['Player'].count()
# for i in range(len(cnt)):
#   fig.annotate(i,cnt[i]/2,str(cnt[i]),fontdict=dict(color='White',fongtsize=12),va='center')
```

```python
ipl.groupby('status')['Player'].count()
```

```python
# Total no. of players bought by each team
plt.figure(figsize=(20,10))
fig=sns.countplot(x=ipl[ipl['Team']!='Unsold']['Team'])
plt.xlabel('Team names')
plt.ylabel('No. of players')
plt.title('Players bought by each Team',fontsize=12)
plt.xticks(rotation=70)
plt.plot()

for p in fig.patches:
  fig.annotate(format(p.get_height(), '.0f'),(p.get_x()+p.get_width()/2.,p.get_height()-1),ha='center',va='center',xytext=(0,40),textcoords='offset points')
```

```python
ipl['retention']=ipl['Base Price']
ipl['retention'].replace(['2 Cr','40 Lakh','20 Lakh','1 Cr','75 Lakh','50 Lakh','30 Lakh','1.5 Cr'],'From Auction',inplace=True)
```

```python
ipl['Base Price'].replace('Draft Pick',0,inplace=True)
ipl
```

```python
ipl['Base Price Unit']=ipl['Base Price'].apply(lambda x: str(x).split(' ')[-1])
ipl['Base Price Value']=ipl['Base Price'].apply(lambda x: str(x).split(' ')[0])
ipl
```

```python
#Total players retained and bought
ipl.groupby(['Team','retention'])['retention'].count()
```

```python
plt.figure(figsize=(20,10))
fig=sns.countplot(x=ipl[ipl['Team']!='Unsold']['Team'],hue=ipl['TYPE'])
plt.title('Players in each team')
plt.xlabel('Team Names')
plt.ylabel('Numbers in Player')
plt.xticks(rotation=50)
plt.plot()
```

```python
#Highest amt spent on a single player in each team
ipl[ipl['retention']=='From Auction'].groupby('Team')['COST IN ₹ (CR.)'].max().sort_values(ascending=False)
```

```python
ipl[ipl['retention']=='Retained'].sort_values(by='COST IN ₹ (CR.)',ascending=False)
```

```python
# top 5 bowlers
ipl[(ipl['retention']=='From Auction') & (ipl['TYPE']=='BOWLER')].sort_values(by='COST IN ₹ (CR.)',ascending=False)
```
